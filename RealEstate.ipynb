```python
# lower register
# Implicit duplicates

data['locality_name'] = data['locality_name'].str.lower()
data['locality_name'] = data['locality_name'].str.replace('ё', 'е', regex=False)
data['locality_name']
```

```python
# Replace 
# new column

data['only_name'] = data['locality_name'].str.replace(
    r'^(городской посёлок|поселок городского типа|поселок|посёлок|деревня|село|'
    r'садовое товарищество|коттеджный поселок|садоводческое некоммерческое товарищество) ', 
    '',
    regex=True
)

duplicates = data.groupby('only_name')['locality_name'].unique()

potential_duplicates = {}
for k, v in duplicates.items():
    if len(v) > 1:
        potential_duplicates[k]= v

for key, val in potential_duplicates.items():
    print(f'{key}: {val}')
```

```python
# Clean locality_name

def clean_l(d, column='locality_name'):
    d[column] = (
        d[column]
        .str.replace(r'\bдеревня\b', 'поселок', regex=True)  
        .str.replace(r'\bсело\b', 'поселок', regex=True)  
        .str.replace(r'\bсадовое товарищество\b', 'поселок', regex=True)  
        .str.replace(r'\bпоселок городского типа\b', 'поселок', regex=True)  
        .str.strip() 
    )
    return d

data = clean_l(data)
data.head().T
```

## Analysis and processing of missing values 1

### Reasons for missing values

### 1. `ceiling_height`.  
**Reasons for omissions:**  
- Sellers often do not specify ceiling heights because it is not always considered an important parameter.  
- Some homes have standardized ceiling heights and the seller may not have felt it necessary to fill it in.  
- The data source may not have included ceiling heights for all listings.  

### 2. `floors_total` (total floors)  
**Reasons for omissions:**  
- Errors or inaccuracies in the data source.  
- Some listings may have been uploaded with incomplete information.  
- Some data may have been lost in processing or uploading.  

### 3. `living_area` (living_area).  
**Reasons for omissions:**.  
- Not all sellers know the exact living area, especially in older homes.  
- Data may not have been available from the source, for example if the square footage is approximate.  
- There may have been errors in filling in or uploading the data.  

### 4. `is_apartment`  
**Reasons for omissions:**  
- Many sellers do not specify this parameter because it is standardly assumed that the dwelling is an apartment.  
- In the database, this parameter may not have been initially available for most listings.  
- The omissions may have occurred due to incorrect merging of data from different sources.  

### 5. `kitchen_area` (kitchen area)  
**Reasons for omissions:**.  
- Sellers often only list the total and living area without filling in the kitchen data.  
- Perhaps the kitchen is combined with the
- Errors in data filling or technical problems with uploading information.  

### 6. `balcony` (number of balconies)  
**Reasons for omission:**  
- If the seller did not specify this parameter, there is probably no balcony.  
- Some listings may have been uploaded without balcony information.  
- Data may not have been captured in the source database.  

### 7. `locality_name` (locality)  
**Causes of omissions:**  
- Errors in data filling by vendors.  
- Data may have been lost when combining information from different sources.  
- In rare cases, there may be errors when uploading information.  

### 8. `airports_nearest`, `cityCenters_nearest`, `parks_nearest`, `ponds_nearest`.  
**Reasons for omissions:**  
- If an object is too far from the specified points, the distance may not have been calculated.  
- There may have been errors in geodata or missing coordinates for some listings.  
- The data source may have contained incomplete information for these parameters.  

### 9. `days_exposition` (time on sale)  
**Reasons for omissions:**  
- If a listing was manually unpublished or deleted due to errors, the time it was posted may not have been retained.  
- Data may have been missing from the original upload if there was no information about the date of removal.  
- Errors in the collection or processing of information.  

### Analyzing and handling missing values 2
### 1. Determining if there are any missing values  
- Count the number of missing values in each column.  
- Calculate the percentage of skips from the total number of rows.  

### 2. Analyze each column with skips  
For each of the following columns, identify possible causes of skips and methods of treatment:

#### `ceiling_height` (9195 skips ≈ 38.8%)  
- Let's check the distribution of values.  
- Identify typical values of ceiling heights and replace omissions with the median or most frequent value.  

#### `floors_total` (86 skips ≈ 0.4%)  
- Let's check if there are skips in `floor`, and if they are in the same rows, it may be worth removing such rows.  
- We can replace by median or mod for similar apartments (e.g. by `locality_name`).  

#### `living_area` (1903 skips ≈ 8%)  
- Let's find the `living_area / total_area` ratio for the filled rows and try to recover the missing values by total area.  

#### `is_apartment` (20924 skips ≈ 88%)  
- Most likely the skips mean that the object **is not an apartment**. Replace `NaN` with `False`.  

#### `kitchen_area` (2278 skips ≈ 9.6%)  
- Let's calculate the median ratio `kitchen_area / total_area` and fill in the skips using the formula `total_area * median_value`, removing the studios from the formula.  

#### `balcony` (11519 skips ≈ 48.6%)  
- It is logical to assume that if there is no data, there is no balcony. Let's replace `NaN` with `0`.  

#### `locality_name` (49 skips ≈ 0.2%)  
- Let's look at examples of data with skips.  
- If there are few of these, we can delete rows or replace `NaN` with ``unknown``.  

#### `airports_nearest`, `cityCenters_nearest`, `parks_around3000`, `parks_nearest`, `ponds_around3000`, `ponds_nearest`.  
- The omissions may mean that the object is far away from these objects.  
- We can leave `NaN` as it is an important signal, we will decide how to use it later.

#### `days_exposition` (3181 skips ≈ 13.4%)  
- Consider the median or mean value for occupancy depending on the presence of outliers. Perhaps it's to important to fill it, maybe we should leave them blank.

```python
#Realize missing values in the dataset
# ceiling_height          9195
# filter height
data.loc[(data['ceiling_height'] < 2.2) | (data['ceiling_height'] > 4), 'ceiling_height'] = np.nan

loc_ceil_median = data.groupby('locality_name')['ceiling_height'].median().to_dict()

# func for filling NaN
def fill_ceil(row):
    if pd.isna(row['ceiling_height']):
        med_val = loc_ceil_median.get('ceiling_height', np.nan)
        return med_val if not pd.isna(med_val) else data['ceiling_height'].median()
    return row['ceiling_height']

data['ceiling_height'] = data.apply(fill_ceil, axis=1)
data['ceiling_height'].describe().T
```

```python
# Delete missing values in floors_total              86

data.dropna(subset=['floors_total'], inplace=True)
```

```python
# Realize missing values in the dataset 
# living_area             1903

living_ratio = (data['living_area'] / data['total_area']).median()
data['living_area'] = data['living_area'].fillna(data['total_area'] * living_ratio)
```

```python
# Realize missing values in the dataset 
# is_apartment           20924
data['is_apartment'].unique()
```

```python
# is_apartment This field contains true and false values, 
# the vast majority of values are not populated, 
# apparently you can delete the field
data = data.drop(columns=['is_apartment'])
```

```python
# Realize missing values in the dataset 
# kitchen_area            2278
# Exclude studios
kitchen_ratio = (
    data.loc[data['studio'] == False, 'kitchen_area'] /
    data.loc[data['studio'] == False, 'total_area']
).median()

data.loc[(data['kitchen_area'].isnull()) & (data['studio'] == False), 'kitchen_area'] = \
    np.minimum(data['total_area'] * kitchen_ratio, data['total_area'] * 0.3)

# Set 0 in studios' kitchen
data.loc[data['studio'] == True, 'kitchen_area'] = 0
```

```python
# Realize missing values in the dataset 
# balcony                11519
data['balcony'] = data['balcony'].fillna(0)
```

```python
# Realize missing values in the dataset 
# locality_name             49
# only_name                 48

data = data.dropna(subset=['locality_name'])
data = data.drop(columns=['only_name'])
```

```python
# Realize missing values in the dataset 
# airports_nearest        5542
# cityCenters_nearest     5519

grouped_median = data.groupby('locality_name')[['airports_nearest',
                                                'cityCenters_nearest']].median()
def fill_nearest(row, column):
    if pd.isna(row[column]):
        if row[column] in grouped_median.index:
            median_value = grouped_median.loc[row['locality_name'], column]
            return median_value if not pd.isna(median_value) else np.nan
    return row[column]

for column in ['airports_nearest', 'cityCenters_nearest']:
    data[column] = data.apply(lambda row: fill_nearest(row, column), axis=1)
```

```python
# Realize missing values in the dataset 
# If the nearest pond and park within a 3km radius isn't full, it's zero
# parks_around3000        5518
# ponds_around3000        5518

data['parks_around3000'] = data['parks_around3000'].fillna(0)
data['ponds_around3000'] = data['ponds_around3000'].fillna(0)
```

```python
# Realize missing values in the dataset 
# Let's first check if the values in the parks and ponds columns are up to 3 km range
# parks_nearest          15620
# ponds_nearest          14589

parks_median = data.groupby('locality_name')['parks_nearest'].median()
parks_median = data.groupby('locality_name')['ponds_nearest'].median()

data.loc[(data['parks_nearest'].isna()) & (data['parks_nearest'] > 0), 'parks_nearest'] = \
    data['locality_name'].map(parks_median)
data.loc[(data['ponds_nearest'].isna()) & (data['ponds_nearest'] > 0), 'ponds_nearest'] = \
    data['locality_name'].map(parks_median)
```
